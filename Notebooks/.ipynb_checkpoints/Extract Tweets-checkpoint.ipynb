```python
import pandas as pd
import json
```

```python
with open('./-f', 'r') as f:
    data = []
    for line in f.readlines():
        data.append(json.loads(line))
```

```python
# This loads the most comprehensive text portion of the tweet  
# Where "tweet" is an individual tweet, treated as JSON / dict
def getText(tweet):       
# Try for extended text of original tweet, if RT'd (streamer)
    try:
        text = tweet['retweeted_status']['extended_tweet']['full_text']
    except: 
    # Try for extended text of an original tweet, if RT'd (REST API)
        try: 
            text = tweet['retweeted_status']['full_text']
        except:
        # Try for extended text of an original tweet (streamer)
            try: 
                text = tweet['extended_tweet']['full_text']
            except:
            # Try for extended text of an original tweet (REST API)
                try: 
                    text = tweet['full_text']
                except:
                # Try for basic text of original tweet if RT'd 
                    try: 
                        text = tweet['retweeted_status']['text']
                    except:
                    # Try for basic text of an original tweet
                        try: 
                            text = tweet['text']
                        except: 
                        # Nothing left to check for
                            text = ''
    return text
```

```python
all_info = pd.DataFrame(data)    
```

```python
tweet_lst = []

for i,item in enumerate(data):
    full_text = {'full_text':getText(item)}
    tweet_lst.append(full_text)
```

```python
all_info['full_text'] = pd.DataFrame(tweet_lst)
all_info.head()
```

```python
all_info.to_csv('./Data/tweets_September30.csv')
```
