# SVD

The main goal of dimensionality reduction is to reduce high-dimensional data into a lower-dimensional subspace while keeping the essential features of the original data as much as possible.  Truncated Singular Value Decomposition (TSVD) is a matrix decomposition method that can be used for feature reduction.  I had more than 60000 features before setting the max_features parameter in my tfidf to 10000.  I will use SVD to further reduce my features to a smaller subset that are most relevant to my prediction problem.

```python
tfidf = TfidfVectorizer(stop_words = stop_word_list,
                        max_df=0.7,
                        max_features=10000)
```

```python
term_mat = tfidf.fit_transform(X_train)
```

```python
term_mat.shape
```

```python
term_df = pd.SparseDataFrame(term_mat, columns=tfidf.get_feature_names())
```

```python
term_df.fillna(0, inplace=True)
```

### Compute SVD

```python
%%time

SVD = TruncatedSVD(n_components=5000)
svd_matrix = SVD.fit_transform(term_df)
svd_matrix.shape

#create svd object with 6,500 components
```

```python
term_df.shape
```

### Explained Variance

Because TSVD provides the ratio of the original feature matrixâ€™s variance explained by each component, I can select the number of components that explain a desired amount of variance.  I have decided on an SVD of 5000 components which would capture 80% of the variance.

```python
fig = plt.figure(figsize=(20,10))

plt.plot(np.array(range(5000))+1, 
         np.cumsum(SVD.explained_variance_ratio_), 
         label='cumulative explained variance')

plt.bar(np.array(range(5000))+1, 
        SVD.explained_variance_ratio_, 
        color='g', 
        label='explained variance')

plt.legend(fontsize=16)
plt.xlabel('component', fontsize=20)
plt.ylabel('variance ratio', fontsize=20)
plt.title('Explained variance by component', fontsize=36);
```

### Convert Matrix to DataFrame

```python
component_names = ["component_"+str(i+1) for i in range(5000)]
svd_df = pd.DataFrame(svd_matrix,
                      columns=component_names)
```

```python
svd_df.shape, len(y_train)
```

```python
svd_df.head()
```

```python
# Must transform the X_test since we fit_transform X_train
X_test_svd = SVD.transform(tfidf.transform(X_test))
```

### Feature Loadings Matrix

```python
loadings = pd.DataFrame(SVD.components_,
                        index=component_names,
                        columns=term_df.columns).T
```

```python
loadings.head()
```

```python
loadings['abs_component_1'] = np.abs(loadings.component_1)
loadings['abs_component_2'] = np.abs(loadings.component_2)
```

```python
loadings.sort_values('abs_component_1',ascending=False).head(20)[['component_1']]
```

```python
loadings.sort_values('abs_component_2',ascending=False).head(20)[['component_2']]
```

### Plot First Two Components

```python
def get_loadings_label(comp_num):
    return ', '.join([*loadings.iloc[np.argsort(np.asarray(np.abs(loadings[f'component_{comp_num}'])))[:-6:-1], :].index])    
```

```python
svd_df['label'] = df_train['label']
svd_df.head()
```

```python
def plot_components(svd_df, first_comp_num, second_comp_num):

    plt.figure(figsize=(20,10))
    x_1 = svd_df[f'component_{first_comp_num}'][svd_df['label']=='pos']
    y_1 = svd_df[f'component_{second_comp_num}'][svd_df['label']=='pos']
    
    plt.scatter(x_1, y_1, 
                c='goldenrod', 
                alpha=.3)
    
    plt.scatter(svd_df[f'component_{first_comp_num}'][svd_df['label']=='neg'],
                svd_df[f'component_{second_comp_num}'][svd_df['label']=='neg'],
                c='grey',
                alpha=.3)

    plt.xlabel(f"{get_loadings_label(first_comp_num)}", fontsize=20)
    plt.ylabel(f"{get_loadings_label(second_comp_num)}", fontsize=20)
    plt.title(f'Component {first_comp_num} vs. {second_comp_num} of our SVD', fontsize=36)
    plt.legend(['Positive', 'Negative'], fontsize=16)
```

```python
plot_components(svd_df, 1,2)
```

This plot shows that there is a lot of overlap between the first and second components.  Ideally, I would like there to be a clear distinction between components or at least some kind of separation between the two.  This demonstrates that many of the same words are used across all components.

### Feature Importance

I will fit a basic logistic regression model with tfidf in order to highlight the most important words and their coefficient weights.  I cannot do this using the sparse data frame that was produced during the svd process.  

```python
pipe = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("lr", LogisticRegression())
])

params = {
    'tfidf__max_df' : [0.7],
    'tfidf__max_features': [10000],
    'lr__penalty': ['l1', 'l2'],
    'lr__C': [0.5, 1.0, 2.0]
}

gs_lr_tfidf = GridSearchCV(pipe, param_grid=params)
gs_lr_tfidf.fit(X_train, y_train);
```

```python
coefs_df = pd.DataFrame(gs_lr_tfidf.best_estimator_.named_steps.lr.coef_, 
                    columns = gs_lr_tfidf.best_estimator_.named_steps.tfidf.get_feature_names()).T
```

```python
coefs_df.columns = ['weights']
coefs_df['abs_w'] = coefs_df.weights.abs()
coefs_df_0 = gs_lr_tfidf.best_estimator_.named_steps.lr.intercept_[0]
weights = coefs_df.sort_values('abs_w', ascending=False)['weights'].head(20)
labels = weights.index
```

```python
plt.figure(figsize=(20,15))
plt.barh(weights.index, weights)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Weights', fontsize=30)
plt.title(f'Logistic Regression Coefficients - Movie Review Classification', fontsize=35)
plt.tight_layout();
```

This plot shows the top 20 words with the highest coefficient weights in either direction of a grid searched logistic regression model with tfidf. Words with negative coefficients contributed heavily to negative reviews, while words with positive coefficients are associated with positive reviews.

### Confusion Matrix

```python
X_test = np.array(X_test).reshape((1, -1))

y_pred = gs_tfidf_svd_lr.best_estimator_.predict(X_test_svd)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
sns.set(font_scale=1.4)
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='g',
           yticklabels = ['Neg. Actual', 'Pos. Actual'],
           xticklabels = ['Neg. Guess', 'Pos. Guess'],
           annot_kws={"size": 16},)

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print("True Negatives: %s" % tn)
print("False Positives: %s" % fp)
print("False Negatives: %s" % fn)
print("True Positives: %s" % tp)
```

```python
print(classification_report(y_test, y_pred))
```

### ROC-AUC Score

```python
proba_pairs = gs_tfidf_svd_lr.best_estimator_.predict_proba(X_test_svd)
gs_tfidf_svd_lr_probas = [item[1] for item in proba_pairs]
roc_auc_score(y_test, gs_tfidf_svd_lr_probas)
```

The Receiver Operating Curve (ROC) plots the true positive rate against the false positive rate. The false positive rate in this case would be the reviews predicted as positives that are actually negative and the false negatives would be the reviews predicted as negatives that were actually positive.  The ROC gives a sense about the tradeoff between sensitivity and specificity; a classifier that predicts at random will appear as the diagonal line. The closer the curve is to the upper left, the better overall accuracy of the model.

This model performs very well in that it accurately represents the tradeoff in which the higher the true positive rate, the more false positives the classifier produces. This is demonstrated by the ROC Curve as it is very close to the upper left corner and its area under the curve (AUC) is 0.99, where a perfect classifier will have a ROC-AUC of 1.0.

```python
fpr, tpr, threshold = roc_curve(y_test, gs_tfidf_svd_lr_probas)
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic', fontsize=20)
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right', fontsize=14)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title("ROC")
plt.legend(loc = 'lower right')
plt.show()
```

While tfidf considers the relative frequency or rareness of tokens in a document against their frequency in other documents, Doc2vec offers an entirely different algorithm than tf-idf in that it represents each document by a dense vector which is trained to predict words in the document.
